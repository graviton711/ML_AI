```python
#Notebook 1/11/2025
#Time: 7:00AM
#Covariance Matrix

import numpy as np

def cov(A):
    N = A.shape[0]
    mean_vector = np.mean(A, axis=0) #axis = 0 calculate mean by column
    print("Mean value by column: ")
    print(mean_vector)
    B = A - mean_vector
    print("Mean subtract: ")
    print(B)

    #@ is dot product
    #B.T is B^T
    covariance_matrix_manual = (B.T @ B) / (N - 1)
    return covariance_matrix_manual

A = np.random.rand(10, 3)
#rowvar=False to calculate the covariance by column, default is by row
covariance_matrix_numpy = np.cov(A, rowvar=False)
covariance_matrix_manual = cov(A)

print("Covariance matrix by numpy: ")
print(covariance_matrix_numpy)
print("Covariance matrix by manual: ")
print(covariance_matrix_manual)


print("Two covariance matrix are equal: ", np.allclose(covariance_matrix_manual, covariance_matrix_numpy))
```

```python
#Notebook 2/11/2025
#Time: 6:43AM
#MSE (Mean Squared Error)
import numpy as np

def compute_cost(X, w, y):
    m = X.shape[0]
    #residual (The distance between model's predicted label and actual label)
    r = (X @ w - y)
    cost = np.sum(r**2) / m
    return cost

#m samples
#n features

m = 10
n = 2

X = np.random.rand(m, n)

#Add one more column for bias
X = np.concatenate((np.ones((m, 1)), X), axis=1)

print("Matrix X: ")
print(X)

#weights vector
w = np.random.rand(n+1, 1)

print("Model weights: ")
print(w)

#Actual label
y = np.random.rand(m, 1)
print("Actual label: ")
print(y)

print("MSE cost: ")
print(compute_cost(X, w, y))
```

```python
#Notebook 2/11/2025
#Time: 9:13AM
#Gradient Descent
import numpy as np

def compute_cost(X, w, y):
    m = X.shape[0]
    r = (X @ w - y)
    cost = np.sum(r**2) / m
    return cost

def gradient_descent(X, w, y, learning_rates, iterations):
    m = X.shape[0]
    cost_history = []

    #iterations: loop time
    for i in range(iterations):
        y_hat = X @ w

        #distances between predicted label and actual label
        errors = y_hat - y

        gradient = (2 / m) * (X.T @ errors)

        #decrease model weights by a small amount
        w = w - (learning_rates * gradient)

        if (i % 100 == 0):
            cost = compute_cost(X, w, y)
            cost_history.append(cost)

    return w, cost_history

m = 10 #samples
n = 2  #features
X = np.random.rand(m, n)
X = np.concatenate((np.ones((m, 1)), X), axis=1)
print("Matrix X: ")
print(X)

w = np.random.rand(n+1, 1)
y = np.random.rand(m, 1)

print("Model weights: ")
print(w)
print("Actual label: ")
print(y)

w, cost_history = gradient_descent(X, w, y, 0.01, 500)

print("Model final weight: ")
print(w)

print("Cost history: ")
print(cost_history)

```

```python
#Notebook 3/11/2025
#Time: 10:16PM
#Linear independent

import numpy as np

def is_linear_independent(vectors):
    num_vec = len(vectors)
    matrix = np.stack(vectors, axis=1)
    print("Convert vectors to matrix: ")
    print(matrix)
    rank = np.linalg.matrix_rank(matrix)
    print("Rank of matrix: ", rank)
    if (rank==num_vec):
        return True
    return False

vectors = []
num_vec = 4

for i in range(num_vec):
    vec = np.random.randint(1, 100, size=(4))
    vectors.append(vec)

print("List of vectors: ")
print(vectors)

print("These vectors are linear independent: ", is_linear_independent(vectors))
```

```python
#Notebook 7/11/2025
#Time: 6:52PM
#compute determinant

import numpy as np

def eliminate_ri_cj(A, i, j):
    B = []
    m = A.shape[0]
    n = A.shape[1]
    for row in range(0, m):
        new_row = []
        for col in range(0, n):
            if (row==i or col==j): continue
            new_row.append(A[row][col])
        if (len(new_row)!=0):
            B.append(new_row)

    new_matrix = np.array(B, dtype=np.int32)
    return new_matrix

def compute_determinant(A, by_row=0):
    #Matrix size m x n
    n = A.shape[1]

    if (n==2):
        return A[0][0] * A[1][1] - A[0][1] * A[1][0]

    determinant = 0
    for i in range(0, n):
        if (A[by_row][i]==0): continue
        determinant+=(-1)**(by_row + i) * A[by_row][i]*compute_determinant(eliminate_ri_cj(A, by_row, i), by_row)
    return determinant


A = np.random.randint(1, 10, size=(4, 4))
print("Matrix A: ")
print(A)
print("Determinant of A: ")
print(compute_determinant(A, 0))
```

```python
#Notebook 8/11/2025
#Time: 12:32PM
#Compute deriative

def f(x):
    return x**2

def compute_deriative(x):
    h = 1e-6
    df_dx = (f(x+h) - f(x)) / h
    return df_dx


x = 2
print("Deriative of f(x)=x^2 at",x,"is:", compute_deriative(x))
```
